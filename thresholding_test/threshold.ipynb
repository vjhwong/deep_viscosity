```python
import matplotlib.pyplot as plt
import cv2
```

```python
# Load the image
image = cv2.imread('blue_vattenkanna.png')

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.axis('off')
```

```python
red_channel = image.copy()[:, :, 0]
green_channel = image.copy()[:, :, 1]
blue_channel = image.copy()[:, :, 2]
fig, axes = plt.subplots(1, 3, figsize=(12, 6))

# Plot red channel
axes[0].imshow(red_channel, cmap='gray')
axes[0].set_title('red channel')
axes[0].axis('off')

# Plot green channel
axes[1].imshow(green_channel, cmap='gray')
axes[1].set_title('green channel')
axes[1].axis('off')

# Plot blue channel
axes[2].imshow(blue_channel, cmap='gray')
axes[2].set_title('blue channel')
axes[2].axis('off')

plt.show()
```

```python
img = red_channel*0.5 + green_channel*0.5 - blue_channel
_, thresh = cv2.threshold(img, 5, 255, cv2.THRESH_BINARY)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))
# Plot red channel
axes[0].imshow(img, cmap='gray')
axes[0].set_title('Image')
axes[0].axis('off')

axes[1].imshow(thresh, cmap='gray')
axes[1].set_title('Thresholded Image')
axes[1].axis('off')

plt.show()
```
