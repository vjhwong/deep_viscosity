```python
from deep_viscosity.utils.rename_data_files import find_interpolated_viscosities
import matplotlib.pyplot as plt
import numpy as np
```

```python
percentages_all, viscosities_all = find_interpolated_viscosities()
print(percentages_all)
viscosities_all = viscosities_all[:-2]
print(len(viscosities_all))
plt.plot(viscosities_all, np.zeros(len(viscosities_all)), 'x')
plt.axvline(x=200, color='r', linestyle='--')
count_below_200 = sum(1 for v in viscosities_all if v < 200)
plt.text(0, 0.5, f'Count below 200: {count_below_200}', color='blue')
plt.axvline(x=600, color='r', linestyle='--')
plt.grid(True)
```
